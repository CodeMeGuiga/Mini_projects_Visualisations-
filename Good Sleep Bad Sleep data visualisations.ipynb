```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_path ="Sleep_Prediction_Dataset.csv"
df = pd.read_csv(data_path)
```

```python
sns.pairplot(df, hue="Good_Sleep", palette="coolwarm")
plt.title("Pairplot of Sleep Data")
plt.show()
```

```python
plt.figure(figsize=(10,6))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation heatmap")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x="Mood_Score", hue="Good_Sleep", kde=True, palette="coolwarm", bins=10)
plt.title("Distribution of Mood Scores by Sleep Quality")
plt.xlabel("Mood Score")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Good_Sleep", y="Rest_Time_Hours", palette="coolwarm")
plt.title("Boxplot of Rest Time by Sleep quality")
plt.xlabel("Good Sleep (1=Yes, 0=No)")
plt.ylabel("Rest Time(Hours)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="Stress_Level", y="Caffeine_Intake_Mg", hue="Good_Sleep", palette="coolwarm", s=50)
plt.title("Stress Level vs. Caffeine Intake")
plt.xlabel("Stress Level")
plt.ylabel("Caffeine Intake (mg)")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Good_Sleep", palette="coolwarm")
plt.title("Count of Good Sleep vs. No Good Sleep")
plt.xlabel("Good Sleep (1=Yes, 0=No)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(data=df, x="Good_Sleep", y="Movement_Hours", palette="coolwarm", split=True)
plt.title("Violin Plot of Movement Hours by Sleep Quality")
plt.xlabel("Good Sleep (1=Yes, 0=No)")
plt.ylabel("Movement Hours")
plt.show()
```

```python
avg_mood_by_stress = df.groupby("Stress_Level")['Mood_Score'].mean().reset_index()
plt.figure(figsize=(8,5))
sns.lineplot(data=avg_mood_by_stress, x="Stress_Level", y="Mood_Score", marker="o", color="blue")
plt.title("Average Mood Score Across Stress Levels")
plt.xlabel("Stress Level")
plt.ylabel("Average Mood Score")
plt.show()
                                
```

```python
sns.jointplot(data=df, x="Rest_Time_Hours", y="Mood_Score", hue="Good_Sleep", kind="kde", palette="coolwarm")
plt.suptitle("Joinplot of Rest Time and Mood Score", y=1.02)
plt.show()
```

```python
avg_caffeine_by_sleep = df.groupby("Good_Sleep")['Caffeine_Intake_Mg'].mean().reset_index()
plt.figure(figsize=(8,5))
sns.barplot(data=avg_caffeine_by_sleep, x="Good_Sleep", y="Caffeine_Intake_Mg", palette="coolwarm")
plt.title("Average Caffeine Intake by Sleep Quality")
plt.xlabel("Good Sleep (1=Yes, 0=No)")
plt.ylabel("Average Caffeine Intake(mg)")
plt.show()
```
